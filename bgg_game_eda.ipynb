```python
import pandas as pd
import numpy as np
import pickle
import ast
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('gameinfo.pkl', 'rb') as read_file:
    game_df = pickle.load(read_file)
```

```python
game_df.head(20)
```

```python
# Finding the number of missing values in each column
game_df.isnull().sum()
```

```python
#Plotting the histogram of rating
plt.hist(game_df["average"])
```

```python
#Plotting the histogram of number of users wanting a game
plt.hist(game_df['wanting'])
```

```python
#Plotting the distribution of number of geeklists
sns.distplot(game_df['numgeeklists'])
plt.title('Average Number of Owners')
plt.show()
```

```python
#Plotting the distribution of game weight (complexity)
sns.distplot(game_df['averageweight'])
plt.title('Average game weight')
plt.show()
```

```python
# Creating new column for average playing time
# Plotting distribution of avg playing time
game_df['avg_time'] = (game_df['minplaytime'] + game_df['maxplaytime']) / 2
avgtime_df = game_df[game_df['avg_time'] < 500]
sns.distplot(avgtime_df['avg_time'], kde=False)
plt.title('Average game time')
plt.show()
```

```python
# Binning game weights
bins = [-1, 1, 2, 3, 4, 5]
game_df['weight_cat'] = pd.cut(game_df['averageweight'], bins=bins, labels=bins[1:])
weight = [game_df[game_df['weight_cat'] == i]['wishing'] for i in range(1,6)]
ax = sns.boxplot(data=weight)
ax.set(xlabel='game weight', ylabel='avgerage\nrating', title='Average Rating of Game Weight Groupings', xticklabels=['0-1', '1-2', '2-3', '3-4', '4-5'])
plt.show()
```

```python
# Binning number of geeklists
owner_bins = [-1, 27, 53, 108, 291, 15269]
game_df['owned_cat'] = pd.cut(game_df['numgeeklists'], bins=owner_bins, labels=range(1,6))
weight = [game_df[game_df['owned_cat'] == i]['average'] for i in range(1,6)]
ax = sns.boxplot(data=weight)
ax.set(xlabel='number of owners', ylabel='avgerage\nowners', title='Average Rating of Game Weight Groupings', xticklabels=['0-1', '1-2', '2-3', '3-4', '4-5'])
plt.show()
```

```python
# creating rating bins
rating_bins = [-1, 5.777, 6.383, 6.956, 10]
game_df['rating_cat'] = pd.cut(game_df['average'], bins=rating_bins, labels=range(1,5))
num_owners = [game_df[game_df['rating_cat'] == i]['numgeeklists'] for i in range(1,5)]
ax = sns.boxplot(data=num_owners)
ax.set(xlabel='Rating Groups', ylabel='avgerage\nowners', title='Average Ownership of Game Weight Groupings', xticklabels=['0-5.7', '5.7-6.3', '6.3-6.9', '6.9+', '4-5'])
plt.show()
```

```python
pd.qcut(game_df['numgeeklists'], q=5)
```

```python
# Visualizing rating per bins of game rankings
num_box = 10
start_vals = range(0, (num_box * 100), 100)
end_vals = [i + 100 for i in start_vals]
weight = [game_df['averageweight'][start:end] for start, end in zip(start_vals, end_vals)]
ax = sns.boxplot(data=weight)
ax.set(xlabel='rank', ylabel='avgerage\nweight', title='Average Weight of Game Ranks',
       xticklabels=['{}-{}'.format(i,j) for i,j in zip(start_vals, end_vals)])
```

```python
# Creating a success metric - conversion rate using number of owners and site views
# Visualizing distribution of conversion rate
game_df['conv_rate'] = (game_df['owned']/game_df['siteviews'])
sns.distplot(game_df['conv_rate'])
```

```python
#  Correlation Matrix
corrmat = game_df.corr()
fig = plt.figure(figsize =(12,9))

sns.heatmap(corrmat, vmax =0.8, square = True)
plt.show()
```

```python
# Pairplot for a selection of non-categorical features
sns.pairplot(data=game_df, vars=['averageweight', 'bayesaverage', 'avg_players','wishing','maxplayers', 'maxplaytime', 'minage', 
                                   'minplayers', 'minplaytime','playingtime'])
plt.title('Correlation plot of selected features for modelling')
plt.show()
```

## Game Growth with Number of Games Available

```python
plt.figure(figsize=(20, 5))
game_df['yearpublished'].value_counts().sort_index().plot()
plt.xlabel('Year')
plt.ylabel('Board Games Published')
plt.title('Number of Board Games Published over Time')
plt.show()
```

```python
game_df['yrs_since_pub'] = 2020 - game_df['yearpublished']
```

```python
with sns.axes_style("darkgrid"):
    game_df['yearpublished'].loc[(game_df['yearpublished'] >= 1900) & (game_df['yearpublished'] < 2019)].value_counts().sort_index().plot(figsize=(10,6), color='tab:orange')
    plt.xlabel('Year')
    plt.ylabel('Board Games Published')
    plt.title('Number of Board Games Published over Time (1900 to 2018)')
    plt.savefig("bgg_yearlyincrease.svg", format="svg")
    plt.show()
```

## Creating Graph Visualizations

```python
# Creating column for average number of players in a game
# Plotting distribution of player counts
plt.figure(figsize=(10, 5))
game_df['avg_players'] = (game_df['minplayers'] + game_df['maxplayers']) / 2
player_counts = game_df['avg_players'].value_counts()
player_counts = player_counts[player_counts > 10] # trim rare occurences
ax = sns.barplot(x=player_counts.index, y = player_counts, color='tab:orange')
plt.tick_params(axis='x', which='major', labelsize=8,rotation=45)
plt.title('Mean Player Counts')
ax.set(ylabel='Number of Games')
ax.set(xlabel='Player Count')
plt.savefig("bgg_playcounts.svg", format="svg", transparent=True)
plt.show()
```

```python
plt.figure(figsize=(20, 7))
board_game_category = game_df['boardgamecategory'].map(lambda artist: ast.literal_eval(artist) if type(artist) != float else artist) \
                                                 .dropna() \
                                                 .sum()
pd.Series(board_game_category).value_counts()[:25].plot(kind='bar', color='darkorange')
plt.rcParams.update({'font.size': 12})
plt.xlabel('Category')
plt.ylabel('Board Game Count')
plt.title('Top 25 Board Game Categories')
plt.savefig('topcategories.svg',format='svg',transparent=True)
plt.show()
```

```python
# removing outliers "Outside Scope of BGG", "Unpublished Prototype", "Miscellaneous Game Accessory"
game_df = game_df[game_df.id != 23953.0]
game_df = game_df[game_df.id != 4249.0]
game_df = game_df[game_df.id != 5985.0]
game_df = game_df[game_df.id != 18291.0]
```

```python
# Plotting top 25 board games by number of owners 
plt.figure(figsize=(20, 7))
game_df.set_index('primary')['owned'].sort_values(ascending=False)[:25].plot(kind='bar', color='tab:orange')
plt.xlabel('Board Game Name')
plt.ylabel('Count')
plt.title('Top 25 Most Owned Board Games (According to Board Game Geek)')
plt.show()
```

```python
# Plotting top 50 designers by number of board games
plt.figure(figsize=(20, 7))
board_game_designer = game_df['boardgamedesigner'].map(lambda artist: ast.literal_eval(artist) if type(artist) != float else artist) \
                                                 .dropna() \
                                                 .sum()
pd.Series(board_game_designer).value_counts()[:50].plot(kind='bar')
plt.xlabel('Board Game Designer')
plt.ylabel('Board Game Count')
plt.title('Top 50 Board Game Designers')
plt.show()
```

```python
# Plotting board game designers for games with rating higher than median
plt.figure(figsize=(20, 7))
board_game_designer = game_df.loc[game_df['bayesaverage'] >= game_df['bayesaverage'].median()]['boardgamedesigner'] \
                               .map(lambda artist: ast.literal_eval(artist) if type(artist) != float else artist) \
                               .dropna() \
                               .sum()
pd.Series(board_game_designer).value_counts()[:50].plot(kind='bar')
plt.xlabel('Board Game Designer')
plt.ylabel('Board Game Count')
plt.title('Top 50 Board Game Designers for Top 50% Board Game Titles based on Bayesian Average')
plt.show()
```

```python
# Plotting top 25 board game mechanics
mechanics = game_df['boardgamemechanic'].map(lambda x: ast.literal_eval(x) if not type(x) == float else x) \
                                          .dropna() \
                                          .sum()
plt.figure(figsize=(20, 7))
pd.Series(mechanics).value_counts()[:25].plot(kind='bar', color='darkorange')
plt.xlabel('Board Game Mechanic')
plt.ylabel('Count')
plt.title('Top 25 Board Game Mechanics')
plt.savefig('topmechanics.svg',format='svg',transparent=True)
plt.show()
```
